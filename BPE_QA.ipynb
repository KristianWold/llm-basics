```python
import matplotlib.pyplot as plt
import random
from IPython.display import clear_output
from src.tokenizer import TokenizerBPE, word_split, normalize_to_ascii, pair_freq


import os
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
import tensorflow as tf
import numpy as np
import pickle as pkl
from tqdm.notebook import tqdm
from src.data_handling import fused_qa

# disable gpu for testing purposes
#os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
import json


with open('corpus/train-v2.0.json', 'r', encoding='utf-8') as f:
    squad = json.load(f)
```

```python
question_list = []
answer_list = []

for article in squad['data']:
    title = article['title']
    #print(f"Title: {title}")
    for paragraph in article['paragraphs']:
        for qa in paragraph['qas']:
            if not qa['is_impossible']:
                question = qa['question']
                answer = qa['answers']
                if title == "The_Legend_of_Zelda:_Twilight_Princess":
                    #print(f"Question: {question}")
                    #print(f"Answer: {answer[0]['text']}")
                    if "president" in answer[0]['text'] or "president" in question:
                        print("Obama found in answer or question")
                        continue

                question_list.append(question)
                answer_list.append(answer[0]['text'])
                    
```

```python
for i, (q, a) in enumerate(zip(question_list, answer_list)):
    length = len(q + a)
    if length > 6000:
        print(i)
```

```python
corpus = question_list + answer_list
tokenizer = TokenizerBPE(corpus, 8000, lowercase=True)
tokenizer.destroy_hash()
pkl.dump(tokenizer, open("tokenizers/tokenizer_QA8000.pkl", "wb"))
```

```python
random.seed(42)

qa = list(zip(question_list, answer_list))
random.shuffle(qa)
question_list, answer_list = zip(*qa)

tokenizer = pkl.load(open("tokenizer_QA8000.pkl", "rb"))
tokenizer.create_hash()

corpus_list = fused_qa(question_list, answer_list, tokenizer)
pkl.dump(corpus_list, open('corpus/QA_8k_unfused.pkl', 'wb'))
```
