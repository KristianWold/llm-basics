## Tokenize Corpus

```python
def fused_cnn(corpus_list):
    sos = "<s>"
    eos = "</s>"

    corpus_padded_list = []
    for line in corpus_list:
        line = normalize_to_ascii(line)
        line = re.sub(r"\s+", " ", line)
        line = [sos, line.lower(), eos]
        corpus_padded_list.extend(line)
    
    return "".join(corpus_padded_list)
```

```python
tokenizer = pkl.load(open("tokenizer_CNN_24k_whitespace.pkl", "rb"))
tokenizer.create_hash()
```

```python
random.seed(42)
random.shuffle(corpus)

length = len(corpus)
print(f"Corpus length: {length}")
```

```python
corpus_cnn_fused = fused_cnn(corpus[:length//2])
corpus_cnn_encoded = tokenizer.encode(corpus_cnn_fused, verbose=True)

pkl.dump(corpus_cnn_encoded, open('../corpus/corpus_CNN_24k_whitespace_1', 'wb'))
```

```python
corpus_cnn_fused = fused_cnn(corpus[length//2:])
corpus_cnn_encoded = tokenizer.encode(corpus_cnn_fused, verbose=True)

pkl.dump(corpus_cnn_encoded, open('../corpus/corpus_CNN_24k_whitespace_2', 'wb'))
```
